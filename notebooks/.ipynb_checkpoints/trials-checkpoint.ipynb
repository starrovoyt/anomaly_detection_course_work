```python
schema = StructType([
    StructField("id1", StringType(), True),
    StructField("id1_type", StringType(), True),
    StructField("id2", StringType(), True),
    StructField("id2_type", StringType(), True),
])


df = ss.read.csv('/Users/starrovoyt/Desktop/hse/course_work/notebooks/data.scv', header=True, schema=schema)
df.printSchema()

df.registerTempTable("soup_edges_table")
```

```python
ss.sql("""
select id1.0 as id_val, id1.1 as id_type, count(distinct(id2)) as neighbours_number
from (
    select map(id1, id1_type) as id1, map(id2, id2_type) as id2
    from soup_edges_table
    union all
    select map(id2, id2_type) as id1, map(id1, id1_type) as id2
    from soup_edges_table
)
group by id1
""").toPandas()
```
